```python
import pandas as pd
import numpy as np

# loading the dataset

df = pd.read_csv('/kaggle/input/job-description-dataset/job_descriptions.csv')
    

# initial info 
print("\nInitial data information:")
print(df.info())


# dropping null and duplicates
relevant_columns = ['Company Size', 'Job Title', 'Experience', 'Country', 
                    'Salary Range', 'Work Type', 'Preference', 'Job Portal', 
                    'Company'] 
df.dropna(subset=relevant_columns, inplace=True)
df.drop_duplicates(inplace=True)
print("\nInitial cleaning (nulls and duplicates) complete.")
print("Number of rows after initial cleaning:", len(df))

# filtering company size
df_filtered = df[df['Company Size'] < 50000].copy()

# filtering job title
df_filtered = df_filtered[df_filtered['Job Title'].str.lower() == 'mechanical engineer']

# filtering experience
df_filtered['Experience'] = df_filtered['Experience'].str.extract('(\d+)').astype(float)
df_filtered = df_filtered[df_filtered['Experience'] > 5]

# filtering salary range
df_filtered['clean_salary'] = df_filtered['Salary Range'].str.replace('$', '').str.replace('k', '000').str.split('-').str[0].astype(float)
df_filtered = df_filtered[df_filtered['clean_salary'] > 50000]

# filtering work type
df_filtered = df_filtered[df_filtered['Work Type'].isin(['Full-time', 'Part-time'])]

# filtering preference
df_filtered = df_filtered[df_filtered['Preference'] == 'Male']

# filtering job portal
df_filtered = df_filtered[df_filtered['Job Portal'].str.lower() == 'idealist']

df_filtered.to_csv('task2_cleaned_data.csv', index=False)
```
