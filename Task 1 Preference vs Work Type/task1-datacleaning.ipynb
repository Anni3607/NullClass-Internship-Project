```python
import pandas as pd

df = pd.read_csv('/kaggle/input/job-description-dataset/job_descriptions.csv')


df = df.drop_duplicates()


df = df.dropna()

filtered_df = df[
    (df['Work Type'] == 'Intern') &
    (df['latitude'] < 10) &
    (~df['Country'].str.startswith(('A', 'B', 'C', 'D'))) &
    (df['Job Title'].str.len() <= 10) &
    (df['Company Size'] != 'Unknown')
]
filtered_df.head()

filtered_df.to_csv("task1_cleaned.csv", index=False)
```
