```python
import pandas as pd


df = pd.read_csv('/kaggle/input/job-description-dataset/job_descriptions.csv')


asian_countries = [
    'Afghanistan', 'Armenia', 'Azerbaijan', 'Bahrain', 'Bangladesh', 'Bhutan', 
    'Brunei', 'Cambodia', 'China', 'Cyprus', 'Georgia', 'India', 'Indonesia', 
    'Iran', 'Iraq', 'Israel', 'Japan', 'Jordan', 'Kazakhstan', 'Kuwait', 'Kyrgyzstan', 
    'Laos', 'Lebanon', 'Malaysia', 'Maldives', 'Mongolia', 'Myanmar (Burma)', 
    'Nepal', 'North Korea', 'Oman', 'Pakistan', 'Palestine', 'Philippines', 
    'Qatar', 'Russia', 'Saudi Arabia', 'Singapore', 'South Korea', 'Sri Lanka', 
    'Syria', 'Taiwan', 'Tajikistan', 'Thailand', 'Timor-Leste', 'Turkey', 
    'Turkmenistan', 'United Arab Emirates', 'Uzbekistan', 'Vietnam', 'Yemen'
]


start_date = '2023-01-01'
end_date = '2023-06-01'


filtered_df_task3 = df[
    (df['Role'] == 'Data Engineer') & 
    (df['Job Title'] == 'Data Scientist') &
    (df['Preference'] == 'Female Only') &
    (df['Qualifications'] == 'B.Tech') &
    (~df['Country'].isin(asian_countries)) & # Exclude Asian countries
    (~df['Country'].str.startswith('C', na=False)) & # Exclude countries starting with 'C'
    (df['latitude'] < 10) &
    (df['Job Posting Date'] >= start_date) & 
    (df['Job Posting Date'] <= end_date)
]

print(f"Number of jobs matching all criteria for Task 3: {len(filtered_df_task3)}")
```

```python
import pandas as pd


df = pd.read_csv('/kaggle/input/job-description-dataset/job_descriptions.csv')


african_countries = [
    'Algeria', 'Angola', 'Benin', 'Botswana', 'Burkina Faso', 'Burundi', 
    'Cabo Verde', 'Cameroon', 'Central African Republic', 'Chad', 'Comoros', 
    'Democratic Republic of the Congo', 'Djibouti', 'Egypt', 'Equatorial Guinea', 
    'Eritrea', 'Eswatini', 'Ethiopia', 'Gabon', 'Gambia', 'Ghana', 'Guinea', 
    'Guinea-Bissau', 'Ivory Coast', 'Kenya', 'Lesotho', 'Liberia', 'Libya', 
    'Madagascar', 'Malawi', 'Mali', 'Mauritania', 'Mauritius', 'Morocco', 
    'Mozambique', 'Namibia', 'Niger', 'Nigeria', 'Rwanda', 'Sao Tome and Principe', 
    'Senegal', 'Seychelles', 'Sierra Leone', 'Somalia', 'South Africa', 
    'South Sudan', 'Sudan', 'Tanzania', 'Togo', 'Tunisia', 'Uganda', 'Zambia', 'Zimbabwe'
]


filtered_df_task4 = df[
    (df['Qualifications'].isin(['B.tech', 'M.tech', 'PhD'])) &
    (df['Work Type'] == 'Full Time') &
    (df['Country'].isin(african_countries)) &
    (df['Job Title'].str.startswith('D', na=False)) &
    (df['Preference'] == 'Male') &
    (df['Company Size'] > 80000) &
    (df['Contact Person'].str.startswith('A', na=False)) &
    (df['Job Portal'] == 'indeed')
]

print(f"Number of jobs matching all criteria for Task 4: {len(filtered_df_task4)}")
```
